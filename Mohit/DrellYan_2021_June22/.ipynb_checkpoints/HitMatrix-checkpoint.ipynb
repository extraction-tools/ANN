```python
import uproot
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
notAHit = 2147483647 #This is the number returned when something is not a hit
```

```python
DY = uproot.open("DrellYan_2021June22_100K.root")
```

```python
ak_arrays = DY[DY.keys()[0]].arrays(filter_name="/H\d[\S]*/", library='pd').replace(notAHit, -1).astype(np.int_)
#replaces the big number that signifies no hits with -1

ak_arrays
# ak_arrays.tolist()
```

```python
per_event = []
x_axis = np.array(ak_arrays.columns)
for (index, subindex), row in ak_arrays.iterrows(): #subindex indicates either first muon or second muon
    if index > 5:
        break
    plt.figure();
    row = np.array(row)
    mask = []
    for n, elementID in enumerate(row):
        if elementID == -1:
            mask.append(False)
        else:
            mask.append(True)
    
    plt.plot(x_axis[mask], row[mask], marker='o');
    plt.tight_layout();
    
```

```python
detector = DY[DY.keys()[0]]["detectorID"].arrays(library='pd').replace(notAHit, -1).astype(np.int_)
for i in detector.columns:
    if detector[i].eq(-1).all():
        detector.drop(i, axis=1,inplace=True)
detector
```

```python
for i in ak_arrays.columns:
    for j in detector.columns:
        if ak_arrays[i] == detector[j]:
            print("Detector with name", i, "Has Detector ID", j[-3:])
            break
```
